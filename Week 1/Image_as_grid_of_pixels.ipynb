```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg  
```

```python
image = mpimg.imread('images/waymo_car.jpg')
print("Image diemtion:", image.shape)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image, cmap='gray')
```

```python
# specific gray scale values
x = 400
y = 390
print(gray_image[y, x])
```

```python
# maximum and minimum gray sace values
max_val = np.amax(gray_image)
min_val = np.amin(gray_image)

print("Maximum values is {} and Minimum value is {}".format(max_val, min_val))
```

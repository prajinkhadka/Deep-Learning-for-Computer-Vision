```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
image = cv2.imread('images/water_balloons.jpg')
```

```python
image_copy = np.copy(image)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
# COLOR CHANNELS
# RGB COLRO SPACE
r = image[:, :, 0]
g = image[:,:, 1]
b = image[:, :, 2]

f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 10))

ax1.set_title("Red channel")
ax1.imshow(r, cmap='gray')

ax2.set_title('Green channel')
ax2.imshow(g, cmap='gray')

ax3.set_title("Blue channel")
ax3.imshow(b, cmap = 'gray')
```

```python
# RGB TO HSV
hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)
plt.imshow(hsv, cmap ='gray')
```

```python
h = hsv[:, :, 0]
s =hsv[:, :, 1]
v = hsv[:, :, 2]



f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20,10))

ax1.set_title('Hue')
ax1.imshow(h, cmap='gray')

ax2.set_title('Saturation')
ax2.imshow(s, cmap='gray')

ax3.set_title('Value')
ax3.imshow(v, cmap='gray')

```

```python
# Pink color 
# Threshold in HSV
lower_hue = np.array([160, 0, 0])
upper_hue = np.array([180, 255 ,255])
```

```python
lower_pink = np.array([180, 0, 0])
upper_pink = np.array([255, 255, 230])
```

```python
# mask the image 
# RGB 
mask_rgb = cv2.inRange(image, lower_pink, upper_pink)
masked_imaged = np.copy(image)
masked_imaged[mask_rgb == 0] = [0, 0, 0]
plt.imshow(masked_imaged)
```

```python
# TRRING HSV
mask_hsv = cv2.inRange(image, lower_hue, upper_hue)

masked_image_hsv= np.copy(image)
masked_image_hsv[mask_hsv==0] = [0, 0, 0]
plt.imshow(masked_image_hsv)
```

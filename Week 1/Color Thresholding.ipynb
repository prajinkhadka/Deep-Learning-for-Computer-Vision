```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
# Reading image 
image = cv2.imread('images/pizza_bluescreen.jpg')
```

```python
print("The image is {} with dimension {}".format(type(image), image.shape))
```

```python
plt.imshow(image)
```

```python
image_copy = np.copy(image)
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)
plt.imshow(image_copy)
```

```python
# Color threshold 
lower_blue = np.array([0, 0, 230])
upper_blue = np.array([50, 50, 255])

# Mask
mask = cv2.inRange(image_copy, lower_blue, upper_blue)
plt.imshow(mask, cmap ='gray')
```

```python
masked_image = np.copy(image_copy)
masked_image[mask!=0] = [0, 0, 0]
plt.imshow(masked_image)
```

```python
#  add background
background_image = cv2.imread('images/space_background.jpg')
background_image = cv2.cvtColor(background_image, cv2.COLOR_BGR2RGB)

# CROP
croped_back = background_image[0:514, 0:816]

# mask
croped_back[mask == 0] =[0, 0, 0]
plt.imshow(croped_back)
```

```python
new_image = croped_back + masked_image
plt.imshow(new_image)
```
